##파이썬 데이터형 : 숫자형

```
# ch2_01_numeric.py
#
print("숫자형: 정수")
a = 123
print(a)

```

```
a = -178
b = 0
a,b
```

```
print("숫자형: 실수")
a = 1.2
b = -3.48
c = 4.24e10
d = 4.24e-10

a,b,c,d
```

```
print("숫자형: 8진수와 16진수")
a = 0o177
b = 0x8FF
c = 0xABC

a,b,c
```

```
print("숫자형: 연산자와 연산")
a = 4
b = 7

a,b
```

```
a + b, a-b, a * b, a/b


# a/b 값이 실수로 나옴
```

```
a ** b, a % b, a // b 

# a**b ---->  a의 b승  --> b가 a의 지수
# a % b ----> 나머지 값
# a // b -----> 몫의 정수값만
```
