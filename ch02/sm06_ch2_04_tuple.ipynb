#튜플(tuple)

**튜플이란?**

리스트는 [ ]으로 둘러싸지만 튜플은 ( )으로 둘러싼다. {}는 딕셔너리와 셋에서 사용. 하지만 인덱스값은 []사용.

리스트는 그 값의 생성, 삭제, 수정이 가능하지만 튜플은 그 값을 바꿀 수 없다!!

```
even = (2, 4, 6, 8)
even
```

여러 가지 튜플

```
t1 = ()  # 튜플 초기화
t2 = (1,) # 튜플이 단일 요소일때는 콤마를 반드시 붙여야 한다.
t3 = (1, 2, 3)
t4 = 1, 2, 3 #튜플의 특징() 생략가능
t5 = ("a", "b", ("ab", "cd"))

print("t1 = ",t1)
print("t2 = ",t2)
print("t3 = ",t3)
print("t4 = ",t4)
print("t5 = ",t5)
```

**튜플의 요소값을 지우거나 변경하려고 하면 어떻게 될까?**

1. 튜플 요솟값을 삭제하려 할 때

```
t1 = (1, 2, "a", "b")
del t1[0]  # Error -----> 튜플은 삭제 불가
```

2. 튜플 요솟값을 변경하려 할 때

```
t1[0] = "c"  # TypeError -----> 튜플은 변경 불가
```

**튜플: 사용방법**

 1.인덱싱하기

```
t1 = (1, 2, "a", "b")
t1
```

```
print("t1[0] = ",t1[0])
print("t1[3] = ",t1[3])
```

2.슬라이싱하기

```
t1 = (1, 2, "a", "b")
t1
```

```
print("t1[1:] = ",t1[1:])
print("t1[2:-1] = ",t1[2:-1])  # ('a',)
print("t1[2:] = ",t1[2:])
```

3.튜플 더하기

```
t1 = (1, 2, "a", "b")
t2 = (3, 4)
t1 + t2

print("t1 = ",t1)
print("t2 = ",t2)
print("t1 + t2 = ",t1 + t2)
```

4.튜플 곱하기

```
t2 = (3, 4)
print("t2 = ",t2)
```

```
id(t2)
```

```
t2 * 3
print("t2 * 3 = ",t2 * 3)
```

```
id(t2)
print("id(t2) = ",id(t2))
```

```
t2 = t2*3
```

```
id(t2) # 주소 달라짐
print("id(t2) = ",id(t2)) 
```

5.튜플 길이 구하기

```
t1 = (1, 2, "a", "b")
print("len(t1) = ",len(t1))
```
