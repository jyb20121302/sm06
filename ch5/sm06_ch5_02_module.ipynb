##모듈이란?

###모듈이란 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일

##모듈 만들기

###간단한 모듈을 한번 만들어 보자.

mod1.py를 현재 폴더에 저장

###파이썬 확장자 .py로 만든 파이썬 파일은 모두 모듈이다.




##모듈 불러오기

###import 모듈명
###import는 이미 만들어 놓은 파이썬 모듈을 사용할 수 있게 해주는 명령어
 

```
# 현재 폴더에 있는 모듈, mod1.py를 연결하기.
# os활용 해서 mod1이 있는 폴더로 변경해야 사용 가능 -> .py 는 쓰면 안됨
import mod1

print(mod1.add(3, 4))

print(mod1.sub(7, 2))
```

```
# 모듈 내부 함수 이용
# from 모듈이름 import 모듈함수

from mod1 import add
add(3, 4)
```

```
# from 모듈이름 import * -> 모든 함수 사용 가능
# from mod1 import add, sub
from mod1 import *

add(3, 4)
sub(7, 3)
```

###if __name__ == "__main__"을 사용하면 C:\doit>python mod1.py처럼 
###직접 이 파일을 실행했을 때는 __name__ == "__main__"이 참이 되어 
###if문 다음 문장이 수행된다. 
###반대로 대화형 인터프리터나 다른 파일에서 이 모듈을 불러서 사용할 때는 
###__ name __ == "__ main __"이 거짓이 되어 if문 다음 문장이 수행되지 않는다.
