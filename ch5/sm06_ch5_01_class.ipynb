##클래스(class)란

###지정된 기능을 수행하는 변수와 함수(메소드)의 모음

##클래스가 필요한 이유

인스턴스, 즉 객체를 생성하여 기능을 이용

코딩을 객체화


```
# [예] 이전에 계산한 결괏값을 기억하고 있는 계산기
result = 0

def add(num):
    global result
    result += num
    return result
```

```
# 이전에 계산한 결괏값을 기억
print(add(3))
print(add(4))
```

```
# 2대의 계산기가 필요한 상황이 발생하면?
result1 = 0
result2 = 0

def add1(num):
    global result1  ## 가능하면 global 사용 X
    result1 += num
    return result1


def add2(num):
    global result2
    result2 += num
    return result2
```

```
print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

###계산기 1의 결괏값이 계산기 2에 아무 영향을 끼치지 않음을 확인할 수 있다. 
###하지만 계산기가 3개, 5개, 10개로 점점 더 많이 필요해진다면 어떻게 해야 할까? 
###그때마다 전역 변수와 함수를 추가할 것인가? 


##클래스를 사용하면 간단하게 해결


```
# class Calculator
# self -> 객체 자신 
class Calculator:
    def __init__(self):    ## init -> 초기화
        self.result = 0


    def add(self, num):
        self.result += num
        return self.result


    def sub(self, num):     ## 함수추가가능
         self.result -= num
         return self.result



cal1 = Calculator()
cal2 = Calculator()
```

```
print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

```
# 뺄셈(sub)을 추가하고 클래스의 객체를 다시 만들어서 사용한다.
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(5))
print(cal1.sub(1))
print(cal2.add(3))
print(cal2.sub(1))
```

###4. 곱하기, 빼기, 나누기 기능 추가

```
# function sub(), mul(), div()

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
        
        
    def add(self):
        result = self.first + self.second
        return result
      

    def sub(self):
        result = self.first - self.second
        return result
      

    def mul(self):
        result = self.first * self.second
        return result
      

    def div(self):
        result = self.first / self.second
        return result
      
```

```
a = FourCal()
b = FourCal()

a.setdata(4, 2)
b.setdata(3, 8)

print(a.add(), a.sub(), a.mul(), a.div())
print(b.add(), b.sub(), b.mul(), b.div())
```

##클래스의 상속
###상속은 기존 클래스를 변경하지 않고 기능을 추가하거나 
###기존 기능을 변경하려고 할 때 사용
###class 클래스 이름(상속할 클래스 이름)

###클래스 FourCal를 상속하는 class MoreFourCal# 
###부모 클래스 FourCal을 가진 자식클래스 MoreFourCal

```
class MoreFourCal(FourCal):
    pass


a = MoreFourCal(4, 2)
a.add(), a.mul()
```

###상속은 MoreFourCal 클래스처럼 기존 클래스(FourCal)는 그대로 놔둔 채 클래스의 기능을 확장시킬 때 주로 사용.
###FourCal 클래스에 a^b (a의 b제곱)을 구할 수 있는 기능을 추가

```
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
    

a = MoreFourCal(4, 2)
print(a.add(), a.mul())
print(a.pow())
```

##메서드 오버라이딩
###기존의 함수(메소드)를 수정/보완하여 재정의 -> 자식클래스가 부모클래스와 다르게 쓸 떄 사용


```
a = FourCal(4, 0)
a.div()  # ZeroDivisionError:
```

```
# 메서드 오버라이딩 (Overriding)
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:  # 나누는 값이 0인 경우 0을 리턴하도록 수정
            print("분모는 0을 사용할 수 없다.")
            return 0
        else:
            return self.first / self.second
    


a = SafeFourCal(4, 0)
a.div()  # 메서드 오버라이딩 사용
```

##클래스 변수
###객체변수와 클래스변수의 차이?
###객체변수   :   객체.변수
###클래스변수 : 클래스.변수


```
class Family:
    lastname = "김"   # 클래스변수
```

```
# 클래스이름.클래스 변수로 사용
Family.lastname
```

```
# 객체변수로 접근
a = Family()
b = Family()

print(a.lastname)
print(b.lastname)
id(a.lastname), id(b.lastname) # 클래스변수는 동일한 주소
```

```
# 클래스변수를 변경하면?
Family.lastname = "박"

print(a.lastname)
print(b.lastname)
id(a.lastname), id(b.lastname), id(Family.lastname)
# 클래스 변수가 공유됨을 확인.
```

```
# 객체변수를 변경하면?
a.lastname = "이"
a.lastname, b.lastname, Family.lastname
id(a.lastname), id(b.lastname), id(Family.lastname)

# 실무 프로그래밍에는 클래스 변수보다는 객체변수를 많이 사용.
```
