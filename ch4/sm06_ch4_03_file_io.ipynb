##파일 읽기

파일 객체 = open(파일 이름, 파일 열기 모드)

파일읽기

readline() 함수 이용하기  --> 한줄 읽기

readlines 함수 사용하기   --> 다 잃기

```
# 1. readline() 함수
# readline_test.py
f = open("새파일.txt", "r")
line = f.readline()
print(line)
f.close()  ## 항상 꼭 닫아줘야한다.
```

```
# readline_all.py
f = open("새파일.txt", "r")
while True:
    line = f.readline()
    if not line:   ### 글자가 없으면 끝내고 출력
        break
    print(line)
f.close()
```

```
# 사용자 입력을 이용한 화면 출력
while 1:
    data = input()
    if not data:   ### 입력이 있으면 출력 없으면 끝냄
        break
    print(data)

```

```
# 2. readlines() 함수
f = open("새파일.txt", "r")
lines = f.readlines()
for line in lines:
    print(line)  # , end="")
f.close()
```

```

# 3. read() 함수
f = open("새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

###파일에 새로운 내용 추가하기

파일에 원래 있던 값을 유지하면서 단지 새로운 값만 추가

파일을 추가 모드('a')로 열고 쓴다.

```
# adddata.py
f = open("새파일.txt", "a")
for i in range(11, 20):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```
# 수정된 파일 확인
f = open("새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

###with문과 함께 사용하기

with문을 사용하면 with 블록을 벗어나는 순간

열린 파일 객체 f가 자동으로 close 된다.

```
f = open("foo.txt", "w")
f.write("Life is too short, you need python")
f.close()
```

```
with open("foo1.txt", "w") as f:
    f.write("Life is too short, you need python")
```
